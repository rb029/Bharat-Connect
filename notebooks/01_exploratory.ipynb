Quick EDA scaffold. Run `python generate_data.py` and `python data_cleaning.py` first.

```python
import pandas as pd, numpy as np
import plotly.express as px

fact = pd.read_parquet('../data/clean/fact_sales.parquet')
fact.head()
```

```python
# Monthly revenue
monthly = fact.groupby('year_month')['line_total'].sum().reset_index()
px.line(monthly, x='year_month', y='line_total', title='Monthly GMV')
```

```python
# Rural vs urban split
split = fact.groupby(fact['is_rural'].map({True:'Rural', False:'Urban'}))['line_total'].sum()
split
```

```python
# Top sellers
fact.groupby('seller_name')['line_total'].sum().nlargest(10)
```
